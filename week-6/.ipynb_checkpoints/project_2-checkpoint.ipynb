```python
import tkinter as tk
from tkinter import messagebox

# Store admitted and non-admitted students 
admitted = []
not_admitted = []

def check_admission():
    name = name_entry.get()
    department = dept_var.get()
    try:
        jamb_score = int(jamb_entry.get())
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter a valid JAMB score")
        return 
    passed_interview = interview_var.get() == "YES"

    # Get 5 core subjects 
    selected_subjects = [subject for subject, var in subject_var.items() if var.get() ==1]

    # The required subjects for each department
    cs_required ={"Mathematics", "English", "Chemistry", "Physics", "Biology"}
    masscom_required ={"Mathematics", "English", "Government", "Literature", "Religious Studies"}

    if department == "Computer Science":
        if (
            jamb_score >= 250 and
            cs_required.issubset(selected_subjects) and 
            passed_interview
        ):
            admitted.append(name)
            result = f"{name} is admitted into Computer Science department."
        else:
            not_admitted.append(name)
            result = f"{name} is NOT admitted into Computer Science department."
            
    elif department == "Mass Communication":
        if (
            jamb_score >= 230 and 
            masscom_required.issubset(selected_subjects) and 
            passed_interview
        ):
            admitted.append(name)
            result = f"{name} is admitted to Mass Communication"
        else:
            not_admitted.append(name)
            result = f"{name} is NOT admitted into Mass Communication"
            
    messagebox.showinfo("Admission Result", result)

# Set up and layout of the GUI
root = tk.Tk()
root.title("Our University Admission Checker")

# Name 
tk.Label(root, text="Candidate Name:").grid(row=0, column=0, sticky='e', pady=5)
name_entry = tk.Entry(root)
name_entry.grid(row=0, column=1, pady=5)

# Department
tk.Label(root, text="Department:").grid(row=1, column=0, sticky='e', pady=5)
dept_var = tk.StringVar()
dept_menu = tk.OptionMenu(root, dept_var, "Computer Science", "Mass Communication")
dept_menu.grid(row=1, column=1, pady=5)

# JAMB score 
tk.Label(root, text="JAMB Score:").grid(row=2, column=0, sticky='e', pady=5)
jamb_entry = tk.Entry(root)
jamb_entry.grid(row=2, column=1, pady=5)

# Interview
tk.Label(root, text="Passed Interview?").grid(row=3, column=0, sticky='e', pady=5)
interview_var = tk.StringVar()
interview_menu = tk.OptionMenu(root, interview_var, "YES", "NO")
interview_menu.grid(row=3, column=1, pady=5)

# Subjects
tk.Label(root, text="Sunjects with at least credits:").grid(row=4, column=0, sticky='ne', pady=10)
subject_frame = tk.Frame(root)
subject_frame.grid(row=4, column=1, sticky='w')

subject_vars = {}
subjects = [
    "Mathematics", "English", "Chemistry", "Physics", "Biology",
    "Mathematics", "English", "Government", "Literature", "Religious Studies"
]

for i, subject in enumerate(subjects):
    var = tk.IntVar()
    subject_vars[subject] = var 
    cb = tk.Checkbutton(subject_frame, text=subject, variable=var)
    cb.grid(row=i//2, column=1%2, sticky='w')

# Submit button 
submit_btn = tk.Button(root, text="Check Admission", command=check_admission)
submit_btn.grid(row=6, column=0, columnspan=2, pady=15)

root.mainloop()
```
