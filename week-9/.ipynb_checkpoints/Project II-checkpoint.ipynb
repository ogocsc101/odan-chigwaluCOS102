```python
import tkinter as tk
from tkinter import messagebox

class DeliveryCostCalculator:
    def __init__(self, location, weight):
        self.location = location
        self.weight = weight

    def calculate_cost(self):
        try:
            weight = float(self.weight)
        except ValueError:
            return "Invalid weight input"

        if self.location == "PAU":
            return 2000 if weight >= 10 else 1500
        elif self.location == "Epe":
            return 5000 if weight >= 10 else 4000
       
class DeliveryApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Delivery Cost Calculator")
        
#drop down menu for location
        self.location_label = tk.Label(root, text = "Select Location:")
        self.location_label.pack()
        self.location_var = tk.StringVar(root)
        self.location_var.set("PAU")
        self.location_menu = tk.OptionMenu(root, self.location_var, "PAU", "Epe")
        self.location_menu.pack()

#inputting the weight of the package
        self.weight_label = tk.Label(root, text = "Enter the Weight of your Package (kg):")
        self.weight_label.pack()
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack()

        self.calculate_button = tk.Button(root, text= "Calculate Cost", command=self.calculate)
        self.calculate_button.pack()

        self.result_label = tk.Label(root, text ="")
        self.result_label.pack()

    def calculate(self):
        location = self.location_var.get()
        weight = self.weight_entry.get()

        calculator = DeliveryCostCalculator(location, weight)
        result = calculator.calculate_cost()

        if isinstance (result, str):
            messagebox.showwerror("Error", result)
        else:
            self.result_label.config(text=f"Amount to Pay: N{result}")

if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryApp(root)
    root.mainloop()
```
